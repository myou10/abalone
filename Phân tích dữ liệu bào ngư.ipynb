Phân tích cho thấy giá trị thuộc tính Rings nằm trong khoảng từ 1 đến 29 vòng trên một mẫu bào ngư.

Có thể thấy rằng có thể xấp xỉ phân phối của thuộc tính này thành một đường cong chuẩn.

Trong phần này, chúng ta phân tích các thuộc tính biểu thị kích thước của một con bào ngư. 
Các thuộc tính này là Chiều dài, Đường kính và Chiều cao. 
Đối với mỗi thuộc tính này, vẽ hai biểu đồ tần suất và biểu đồ hộp tương ứng của chúng.

Một lần nữa, chúng ta quan sát thấy một phân phối chuẩn gần đúng.

```python
#Đọc dữ liệu từ file CSV
abalone = pd.read_csv('abalone.data', names= ['Sex','Length','Diam','Height','Whole','Shucked','Viscera','Shell','Rings'])

# Chỉ lấy những cột có kiểu dữ liệu là số
numeric_df = abalone.select_dtypes(include=['number'])

# Tính toán ma trận tương quan giữa các cột số
corr_matrix = numeric_df.corr()

# Kiểm tra xem có giá trị nào ngoài khoảng -1 đến 1 không
print("Ma trận tương quan:")
print(corr_matrix)
print("Giá trị tối đa và tối thiểu trong ma trận tương quan:")
print("Tối đa:", corr_matrix.max().max())
print("Tối thiểu:", corr_matrix.min().min())
```

## Kết luận
− Có thể xây dựng một mô hình dự đoán giá trị của thuộc tính mục tiêu Rings (số lượng vòng tuổi) thông qua các thuộc tính độc lập như kích thước, trọng lượng và các yếu tố liên quan khác của bào ngư. Mối tương quan giữa các  thuộc tính này cho phép chúng ta sử dụng chúng như các biến dự đoán để
ước lượng tuổi của bào ngư.

− Khối lượng của bào ngư thay đổi theo kích thước: Kích thước và khối lượng của bào ngư có mối quan hệ tương đối chặt chẽ. Khi kích thước bào ngư 
tăng lên, khối lượng của chúng cũng tăng tương ứng. Có thể thấy khi kích  thước lớn hơn thường đi kèm với khối lượng lớn hơn, bao gồm cả thịt, nội 
tạng và vỏ.

− Không có sự khác biệt đáng kể về kích thước, khối lượng và số lượng vòng  (Rings) giữa bào ngư đực và bào ngư cái. Kết quả phân tích cho thấy cả hai 
giới tính có xu hướng phát triển tương tự về mặt sinh học, không có dấu hiệu nào cho thấy một giới tính phát triển nhanh hơn hoặc có số lượng vòng 
tuổi khác biệt so với giới tính còn lại.

− Nhóm bào ngư chưa xác định giới tính (Infant) cho thấy kích thước, khối lượng và số lượng vòng thấp hơn so với nhóm bào ngư trưởng thành. Vì các cá thể sơ sinh chưa phát triển hoàn toàn về mặt kích thước và chưa có đủ số vòng để đạt đến độ tuổi trưởng thành.

− Khối lượng và chiều cao của bào ngư thay đổi đáng kể theo độ tuổi trong giai đoạn phát triển. Tuy nhiên, sau khi bào ngư đạt khoảng 16,5 năm tuổi 
(tương đương với 15 vòng tuổi), kích thước và trọng lượng của chúng dường như không còn thay đổi đáng kể nữa. Điều này có thể cho thấy rằng sau khi đạt đến giai đoạn trưởng thành, sự phát triển của bào ngư về mặt kích thước và trọng lượng gần như dừng lại. Điều này có nghĩa là việc dự đoán tuổi của bào ngư từ các thuộc tính này trở nên khó khăn hơn khi chúng đã trưởng thành hoàn toàn.
